```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = [0.05, 0.15, 0.25, 0.35, 0.45, 0.55, 0.65, 0.75, 0.85]
y = [308.46, 318.43, 327.51, 341.34, 357.38, 382.64, 401.75, 413.65, 424.65]

# Crear el gráfico de dispersión
plt.scatter(x, y)

# Añadir títulos y etiquetas
plt.title('Scatter Plot Cost VS Probability of resources failure')
plt.xlabel('Probability')
plt.ylabel('Cost')

# Calcular la línea de mejor ajuste
coefficients = np.polyfit(x, y, 1)
polynomial = np.poly1d(coefficients)
y_fit = polynomial(x)

# Dibujar la línea de mejor ajuste
plt.plot(x, y_fit, color='blue')

# Mostrar el gráfico
plt.savefig('Cost_Prob_Correlation.pdf', format='pdf')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = [0.05, 0.15, 0.25, 0.35, 0.45, 0.55, 0.65, 0.75, 0.85]
y = [308.46, 318.43, 327.51, 341.34, 357.38, 382.64, 401.75, 413.65, 424.65]

# Crear el gráfico de dispersión
plt.figure(figsize=(5, 3))  # Tamaño pequeño y aplanado

plt.scatter(x, y)

# Añadir títulos y etiquetas con letras grandes
plt.title('Scatter Plot Cost VS Probability of resources failure', fontsize=14)
plt.xlabel('Probability', fontsize=12)
plt.ylabel('Cost', fontsize=12)

# Calcular la línea de mejor ajuste
coefficients = np.polyfit(x, y, 1)
polynomial = np.poly1d(coefficients)
y_fit = polynomial(x)

# Dibujar la línea de mejor ajuste
plt.plot(x, y_fit, color='blue')

# Ajustar tamaño de las marcas y las etiquetas de los ejes
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Mostrar el gráfico y guardarlo
plt.tight_layout()
plt.savefig('Cost_Prob_Correlation.pdf', format='pdf')
plt.show()
```

```python
data = pd.DataFrame({'x': x, 'y': y})
correlation = data['x'].corr(data['y'])
print(correlation)
```

```python
from scipy.stats import t

# Valores de entrada
r = correlation  # Ejemplo del coeficiente de correlación
n = 9   # Número de pares de datos (observaciones)

# Paso 1: Calcular el estadístico t
t_stat = (r * (n - 2)**0.5) / ((1 - r**2)**0.5)

# Paso 2: Determinar los grados de libertad
df = n - 2

# Paso 3: Calcular el valor p
p_value = 2 * (1 - t.cdf(abs(t_stat), df))

print(f'Valor de t: {t_stat}')
print(f'Grados de libertad: {df}')
print(f'Valor de p: {p_value}')
```
